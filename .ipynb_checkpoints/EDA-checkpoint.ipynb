```python
make_ls = train_df['make'].dropna().unique()
make_ls
```

```python
import re


# Combine patterns into a single regex
combined_pattern = r'\b(' + '|'.join(make_ls) + r')\b'

# Precompile the regex
compiled_regex = re.compile(combined_pattern)

# Example string
text = "land rover is the best"

# Use the compiled regex to find all matches
matches = compiled_regex.findall(text)

print("Matches found:", matches)
```

```python
def extract_make(title):
    title = title.lower()
    matches = compiled_regex.findall(title)
    return matches[0]
```

```python
train_df['extracted_make'] = train_df['title'].apply(extract_make)
```

```python
train_df['make'] = train_df['make'].fillna(train_df['extracted_make'])
```

```python
import re

# Example string
text = train_df['features'].iloc[6]
print(text)
# Regex pattern to match engine capacity (floating-point number followed by 'l')
pattern = r'\b(\d+)[ ]?cylinder[s]?\b |\b[v|V][ ]?(\d+)\b'

# Use re.search to extract the engine capacity
match = re.search(pattern, text)

if match:
    print("Number of cylinders:", match.group(1) or match.group(2) )
else:
    print("No match found")
```

```python
import requests

# The URL you want to send the GET request to
url = "https://onemotoring.lta.gov.sg/content/onemotoring/home/buying/upfront-vehicle-costs/open-market-value--omv-.html?year=2003"

# Sending a GET request
response = requests.get(url)

# Checking the status of the request
if response.status_code == 200:
    # Successfully retrieved the content
    print(response.text)  # Output the HTML content of the page
else:
    print(f"Failed to retrieve the content. Status code: {response.status_code}")
```

```python
from bs4 import BeautifulSoup

# Sample HTML content
# html_content = '''
# <div class="brandList" id="brand_OPEL">
#     <input class="brandList_8_2003" type="hidden" id="brandValue_make-19_8_2003" value="OPEL"/>
#     <input class="modelList_make-19_8_2003" type="hidden" id="modelValue_model-0_make-19_8_2003" value="ASTRA AUTO--1598cc"/>
#     <input class="averageOmvList_make-19_8_2003" type="hidden" id="averageOmv_model-0_make-19_8_2003" value="S$19,883"/>
#     <input class="modelList_make-19_8_2003" type="hidden" id="modelValue_model-1_make-19_8_2003" value="VECTRA AUTO--2198cc"/>
#     <input class="averageOmvList_make-19_8_2003" type="hidden" id="averageOmv_model-1_make-19_8_2003" value="S$28,426"/>
#     <input class="modelList_make-19_8_2003" type="hidden" id="modelValue_model-2_make-19_8_2003" value="ZAFIRA AUTO--1796cc"/>
#     <input class="averageOmvList_make-19_8_2003" type="hidden" id="averageOmv_model-2_make-19_8_2003" value="S$22,251"/>
# </div>
# <div class="brandList" id="brand_PERODUA">
#     <input class="brandList_8_2003" type="hidden" id="brandValue_make-20_8_2003" value="PERODUA"/>
#     <input class="modelList_make-20_8_2003" type="hidden" id="modelValue_model-0_make-20_8_2003" value="KELISA EZ A--989cc"/>
#     <input class="averageOmvList_make-20_8_2003" type="hidden" id="averageOmv_model-0_make-20_8_2003" value="S$6,645"/>
#     <input class="modelList_make-20_8_2003" type="hidden" id="modelValue_model-1_make-20_8_2003" value="KELISA GX M--989cc"/>
#     <input class="averageOmvList_make-20_8_2003" type="hidden" id="averageOmv_model-1_make-20_8_2003" value="S$5,951"/>
#     <input class="modelList_make-20_8_2003" type="hidden" id="modelValue_model-2_make-20_8_2003" value="KEMBARA EZ A--1296cc"/>
#     <input class="averageOmvList_make-20_8_2003" type="hidden" id="averageOmv_model-2_make-20_8_2003" value="S$11,815"/>
# </div>
# '''

# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
# Find all year sections (div elements with class 'yearList')
years = soup.find_all('div', class_='yearList')
# Initialize a list to store the extracted data
car_data = []

# Loop through each year and extract the months, brands, models, and values
for year in years:
    year_id = year['id'].split('_')[1]  # Extract year (e.g., "2003" from "year_2003")
    
    # Find all months for the current year
    months = year.find_all('div', class_='monthList')
    
    for month in months:
        month_id = month['id'].split('_')[1]  # Extract month (e.g., "1" from "month_1")
        
        # Find all brands for the current month
        brands = month.find_all('div', class_='brandList')
        
        for brand in brands:
            # Get the brand (make) name
            make = brand.find('input')['value']
            
            # Find all models and values for the current brand
          # Find all models and values for the current brand dynamically
            models = brand.find_all('input', id=lambda x: x and x.startswith('modelValue'))
            values = brand.find_all('input', id=lambda x: x and x.startswith('averageOmv'))
            
            for model, value in zip(models, values):
                model_name = model['value']
                omv_value = value['value']
                # Append a tuple with the year, month, make, model, and value to the list
                car_data.append((year_id, month_id, make, model_name, omv_value))

# Display the extracted car data with year and month
for data in car_data:
    print(f"Year: {data[0]}, Month: {data[1]}, Make: {data[2]}, Model: {data[3]}, OMV: {data[4]}")
```
